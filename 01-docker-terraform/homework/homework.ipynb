## Question 3. Counting short trips

For the trips in November 2025 (lpep_pickup_datetime between '2025-11-01' and '2025-12-01', exclusive of the upper bound), how many trips had a trip_distance of less than or equal to 1 mile?

- 7,853
- 8,007
- 8,254
- 8,421

````python
with engine.connect() as conn:
    res = conn.execute(text("select count(*) from trip_data where lpep_pickup_datetime between '2025-11-01' and '2025-12-01' and trip_distance <= 1"))

    for row in res:
        print(row)
````

A/ `8,007`

## Question 4. Longest trip for each day

Which was the pick up day with the longest trip distance? Only consider trips with trip_distance less than 100 miles (to exclude data errors).

Use the pick up time for your calculations.

- 2025-11-14
- 2025-11-20
- 2025-11-23
- 2025-11-25

````python
with engine.connect() as conn:
    res = conn.execute(text("select lpep_pickup_datetime, trip_distance from trip_data where trip_distance < 100 order by trip_distance desc limit 1"))

    for row in res:
        print(row)
````

## Question 5. Biggest pickup zone

Which was the pickup zone with the largest total_amount (sum of all trips) on November 18th, 2025?

- East Harlem North
- East Harlem South
- Morningside Heights
- Forest Hills


````python
taxizone.dtypes
````

````python
# Create table
taxizone.head(n=0).to_sql(name='zone_data', con=engine, if_exists='replace')
````

````python
taxizone.to_sql(name='zone_data', con=engine, if_exists='append')
````

````python
with engine.connect() as conn:
    res = conn.execute(text("""select sum(total_amount) as total_amount, "Zone" 
                            from trip_data 
                            left join zone_data on "PULocationID" = "LocationID" 
                            where lpep_pickup_datetime between '2025-11-18' and '2025-11-19' 
                            group by "Zone" 
                            order by total_amount desc 
                            limit 1"""))

    for row in res:
        print(row)
````

A/ `East Harlem North`

## Question 6. Largest tip

For the passengers picked up in the zone named "East Harlem North" in November 2025, which was the drop off zone that had the largest tip?

Note: it's tip , not trip. We need the name of the zone, not the ID.

- JFK Airport
- Yorkville West
- East Harlem North
- LaGuardia Airport


````python
with engine.connect() as conn:
    res = conn.execute(text("""select t_d.tip_amount as tip_amount, d_o_z."Zone" as drop_off_zone 
                      from trip_data as t_d 
                      inner join zone_data as p_u_z on t_d."PULocationID" = p_u_z."LocationID"
                      inner join zone_data as d_o_z on t_d."DOLocationID" = d_o_z."LocationID"
                      where t_d.lpep_pickup_datetime between '2025-11-01' and '2025-12-01'
                      and p_u_z."Zone" = 'East Harlem North'
                      order by tip_amount desc limit 1
                      """))
    
    for row in res:
        print(row)
````

A/ `Yorkville West`
